```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import astropy.units as u
import astropy.coordinates as SkyCoord
from astropy.wcs import WCS
```

```python
filename = "WISE_12.fits"
hdu = fits.open(filename)
hdu.info()
header = hdu[0].header

wcs = WCS(header)
wcs
```

```python
#%matplotlib jupyter
ax=plt.subplot(projection=wcs)
ax.imshow(hdu[0].data, origin="lower")
plt.grid()
plt.xlabel("R.A. (deg)")
plt.ylabel("DEC (deg)")
#plt.colorbar()

```
